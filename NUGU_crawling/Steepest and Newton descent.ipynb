```python
import pandas as pd
import numpy as np
from numpy import linalg as la
```

```python
#10.5 (a)

def f(x1,x2,c): # function
    y=c*(x1**2) + (x2**2)
    return y
def f0(x1,x2,c): #first value
    return np.array([[x1],[x2]])
def fp(x1,x2,c): # Gradient
    return np.array([[2*c*x1],[2*x2]])
def dk(x1,x2,c) : # Steepest Descent direction
    return -np.array([[2*c*x1],[2*x2]])
```

```python
c=2
x1=10
x2=10

# 3 Iterations when c=2
for i in range(3):
    for mk in range(0,20):
        xkstep = f0(x1,x2,c)+((0.8)**mk)*dk(x1,x2,c)
        y = f(xkstep[0,0],xkstep[1,0],c) - f(f0(x1,x2,c)[0,0],f0(x1,x2,c)[1,0],c)
        sufdec = 0.5 * (0.8**mk)*(-1)*((2*c*x1)**2+(2*x2)**2)
        if y >= sufdec:
            print(mk," : ",y," > ",sufdec)
        else: break
    print(mk, " : ",y," =< ",sufdec)
    x1 = x1+((0.8)**mk)*dk(x1,x2,c)[0,0]
    x2 = x2+((0.8)**mk)*dk(x1,x2,c)[1,0]
    print ("The result of ",i+1,"th iteration : (x1,x2)= (",x1," , ",x2,")\n")
    mk=0
```

```python
#10.5 (b)

def f(x1,x2,c): # function
    y=c*(x1**2) + (x2**2)
    return y
def f0(x1,x2,c): #first value
    return np.array([[x1],[x2]])
def fp(x1,x2,c): # Gradient
    return np.array([[2*c*x1],[2*x2]])
def dk(x1,x2,c) : # Newton's Method direction
    hessian = np.array([[2*c,0],[0,2]])
    # The Hessian matrix [[2c,0],[0,2]] is pd.
    # Since c is 2, 10, 100 and the leading principal minors are positive.
    return -np.matmul(la.inv(hessian).T,fp(x1,x2,c))
```

```python
c=2
x1=10
x2=10

# 3 Iterations when c=2
for i in range(3):
    for mk in range(0,200):
        xkstep = f0(x1,x2,c)+((0.8)**mk)*dk(x1,x2,c)
        y = f(xkstep[0,0],xkstep[1,0],c) - f(f0(x1,x2,c)[0,0],f0(x1,x2,c)[1,0],c)
        sufdec = 0.5 * (0.8**mk)*float(np.matmul(fp(x1,x2,c).T,dk(x1,x2,c)))
        if y > sufdec:
            print(mk," : ",y," > ",sufdec)
        else: break
    print(mk, " : ",y," =< ",sufdec)
    x1 = x1+((0.8)**mk)*dk(x1,x2,c)[0,0]
    x2 = x2+((0.8)**mk)*dk(x1,x2,c)[1,0]
    print ("The result of ",i+1,"th iteration : (x1,x2)= (",x1," , ",x2,")\n")
    mk=0
```
